It is a fundamental tool for data visualization in Python, offering a wide range of chart types and customization options. Matplotlib is frequently used in data analysis, scientific research, and engineering.

```python
import matplotlib.pyplot as plt  ## Most of the Matplotlib utilities lies under the pyplot submodule.
```

```python
%matplotlib inline   
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x ** 2
```

```python
y
```

```python
## two methods to create plots : Functional and Object Oriented
```

```python
## FUNCTIONaL Method
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
## you can have multiple plots on the same canvas using subplot. -- plt.subplot(nrows, ncols, plot_number)
```

```python
plt.subplot(1,2,1)  ## no of rows, column and plot number
plt.plot(x,y,'g')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
## Object Oriented Method
##The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it. 
```

```python
fig = plt.figure()  ## ths creates a blank object which is used to create plot
axes = fig.add_axes([0.1,0.1,0.8,0.8])   ## define axis for the plot - ## left, up, width & height
axes.plot(x,y)  ## similar plot as earlier
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Plot Title')
```

```python
## Putting 2 figures on one canvas
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])   ## left, up, width & height
axes1.plot(x,y,'g')
axes1.set_title('Big Plot')

axes2.plot(y,x,'r')
axes2.set_title('Small Plot')
```

```python
## Sub-plots using Object Oriented plotting
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)  ## you can see a canvas with 1 row and 2 columns
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)    ## 3 rows and 3 columns. To take care of graphs not clear, use tight_layout method
plt.tight_layout()
```

```python
## 
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
axes  ## axes object is just an array, the one we were creating with axes_add, We can iterate through it.
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for curr_axes in axes:
    curr_axes.plot(x,y)
```

```python
## OR WE CAN EVEN USE THE AXES WITH INDEX NUMBER AND CREATE PLOTS
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('Axes 0')

axes[1].plot(y,x)
axes[1].set_title('Axes 1')
```

```python
## Line Colors and types
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y,color='blue')  ## common colors can be used as string
ax.plot(x,y,color='#800080')  ## RB HEX CODE. Code for purple color here.
```

```python
## Line width and line styles
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#800080',linewidth=4)  ## adding this paramter makes plt line thick or thin depending on the value entered. Can also use lw as shortcut abbrev.
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#800080',linewidth=4,alpha=0.25)  ## adding alpha value makes the line transparent.
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#800080',linewidth=4,alpha=0.50,linestyle='-.')  ## adding linestyle can change the line styles like '--', '..', '-.'.
#possible linestype options -, , -., :, steps
```

```python
## We can also specify the markers which tell us the datapoints on the plt. ## marker and markersize
## # possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#800080',linewidth=4,alpha=0.50,linestyle='-',marker='x',markersize=10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#800080',linewidth=4,alpha=0.50,linestyle='-',marker='o',markersize=15
       ,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='g')
## customizing marker 
```

## Some useful hands-on exercises

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y)
ax2.plot(x,y)
```

```python
## ** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**
## ** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax1.plot(x,z)
ax2.plot(x,y)
ax2.set_title('zoom')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
## ** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**
## ** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**
## ** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(10,2))
axes[0].plot(x,y,linestyle='-',linewidth=4,color='b')
axes[1].plot(x,z,linewidth=2,color='r',ls='--')
```
