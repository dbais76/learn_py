Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps you explore and understand your data. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots. Its dataset-oriented, declarative API lets you focus on what the different elements of your plots mean, rather than on the details of how to draw them.

Behind the scenes, seaborn uses matplotlib to draw its plots. For interactive work, it’s recommended to use a Jupyter/IPython interface in matplotlib mode, or else you’ll have to call matplotlib.pyplot.show() when you want to see the plot.

## Distibution plots

```python
## Visualize distribution of datasets
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')   ## existing demo dataset part of seaborn
```

```python
tips.head()
```

```python
## dist plot -- allows us to plot distribution of uni-variable ie single variable.
```

```python
sns.displot(tips['total_bill'],kde=True)
```

```python
sns.displot(tips['total_bill'],kde=True,bins=30)  ## KDE = Kernel Density Estimation
```

```python
## Joints plot - bi-variable
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)  ## by default it shows a scatter plot. here it shows a relation, higher the bill, higher the tips.
```

```python
## We can chane the type of plot by adding paramter 'kind'.
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')  ## darker color more density. Another example is 'reg' for regression plot & 'KDE' for KDE plot
```

```python
## Pair plot - creates a plot for all numerical columns mapping to each other from the dataset.
```

```python
sns.pairplot(tips)
```

```python
## You can add 'hue' paramter to define category, for ex: in this dataset, male and female
```

```python
sns.pairplot(tips,hue='sex')
```

```python
## We can also define color palette
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
## Rug plot
```

```python
sns.rugplot(tips['total_bill'])  ## this shows dash mark for every single data point, much like histogram
```

```python
##Categorical Plot
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
## counter plot
```

```python
sns.countplot(x='sex',data=tips)
```

```python
## Box plot
```

```python
sns.boxplot(x='day',y='total_bill',data=tips) ## x is categorical data, y is numerical data
```

```python
## Seaborm also allows us to add another categorical to the plot. For ex: smoker in below example added with 'hue'
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
## Violin plot -- very similar to box plot and uses the same inputs
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

```python
## Strip plot  - -sort of scatter plot based on a category
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips) ## It shows more info as it shows density rather than how it is stacked in previous view
```

```python
## We can do 'hue' to add another category.
```

```python
## Swarm plot  -- more of a combination of strip plot and violin plot. It shows all data points.
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
## We can combine violin plot and swarm plot together to show information little clearly.
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='yellow')
```
