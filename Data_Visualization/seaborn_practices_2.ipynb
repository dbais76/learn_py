```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flight = sns.load_dataset('flights')
tips.head()
flight.head()
```

```python
## Matrix plots  - heat map
```

```python
##For heat map or matrx, the data should be in a matrix format, i.e., the row and column should mean something.
```

```python
tips.head()
```

```python
tips.info()
```

```python
## For ex, in this dataset, total_bill variable hold some value, however, index 0,1 etc doesn't show any connection/correlation
```

```python
#  here we can get correlation data
```

```python
## if we don't use (numeric_only=True) with corr() method, it will give error message <could not convert string to float 'no' >, bcoz it is not able to convert string values.
```

```python
tc = tips.corr(numeric_only=True)
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot=True,cmap='coolwarm')  ## annotation and color map added
```

```python
## now to get the flights data in matrix for, we can do a pivot table
```

```python
flight.head()
```

```python
flight.pivot_table(index='month',columns='year',values='passengers')
```

```python
fc = flight.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fc)
```

```python
sns.heatmap(fc,cmap='magma',linecolor='white',linewidths=2) ## making it more legible by adding linewidth and linecolor.
```

```python
## pip install scipy   ## clustermap needs scipy to be installed.
```

```python
sns.clustermap(fc) ## tries to cluster rows and column based on similarities
```

```python
## PairGrid plot is very similar to pairplot, however, it gives more control over what you want to plot and which plot to use
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
## this data set has three species
```

```python
iris['species'].unique()
```

```python
pg = sns.PairGrid(iris)  ## this just creates empty sub-plots allowing us to define what we want.
pg.map_diag(sns.histplot)
pg.map_upper(sns.scatterplot)
pg.map_lower(sns.kdeplot)
```
